```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../data/raw/Superstore_sales.csv",encoding='latin1')

# Show first rows
print(df.head())

# Basic info
print(df.info())

# Total sales by region
region_sales = df.groupby("Region")["Sales"].sum()

# Plot
region_sales.plot(kind="bar")

plt.title("Region Wise Sales")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

```python
import numpy as np
import seaborn as sns
```

```python
# Check missing value 
df.isnull().sum()
```

```python
# Dataset shape
print('Rows','Columns:',df.shape)
```

```python
# statistical summary 
df.describe()
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(category_sales)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=category_sales.index,y=category_sales.values)

plt.title("CATEGORY WISE SALES")
plt.xlabel("CATEGORY")
plt.ylabel("SALES")

plt.show()
```

```python
Top_products = df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(Top_products)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=Top_products.values,y=Top_products.index)
plt.title("Top 10 products by sales")
plt.xlabel("Sales")
plt.ylabel("products")
plt.xticks(color='red')
plt.yticks(fontweight='bold')
plt.show()
```

```python
## Business Insights

#1. Technology category generated highest sales.

#2. West region contributed maximum revenue.

#3. Consumer segment has highest sales contribution.

#4. Some categories have high sales but lower profit margin.

#5. Monthly sales trend shows seasonal growth spikes.
```
