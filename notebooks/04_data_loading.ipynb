```python
#latest order Date

latest_date= df['Order_Date'].max()

#RFM Table

rfm = df.groupby('Customer_ID').agg({
    'Order_Date': lambda x :(latest_date - x.max()).days,
    'Order_ID': 'nunique',
    'Sales': 'sum'
})

#Rename columns

rfm.columns = ['Recency','Frequency','Monetary']

#Display

rfm.head()
```

```python
#RFM Scoring

rfm['R_Score'] = pd.qcut(rfm['Recency'],4,labels=[4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'),4,labels=[1,2,3,4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'],4,labels=[1,2,3,4])

#Total RFM Score
rfm['RFM_Score'] = (rfm['R_Score'].astype(str)+
                    rfm['F_Score'].astype(str)+
                    rfm['M_Score'].astype(str))

#Display
rfm.head()
```

```python
# Customer Segmentation

def customer_segment(rfm_score):

    if rfm_score == '444':
        return 'Champion'

    elif rfm_score[0] == '4':
        return 'Recent Customer'

    elif rfm_score[1] == '4':
        return 'Loyal Customer'

    elif rfm_score[2] == '4':
        return 'Big Spender'

    elif rfm_score == '111':
        return 'Lost Customer'

    else:
        return 'Regular Customer'


# Apply segmentation

rfm['Customer_Segment'] = rfm['RFM_Score'].apply(customer_segment)

# Display

rfm.head(10)
```

```python
#Customer segment count 

segment_count = rfm['Customer_Segment'].value_counts()

#Plot

plt.figure(figsize=(12,5))

segment_count.plot(kind='bar',color=['blue','green','red','orange','purple'])

plt.title('Customer segmentation')
plt.xlabel('segment')
plt.ylabel('Number of Customer')
plt.xticks(rotation=0)
plt.show()
```

```python
#Monthly sales Data

monthly_sales = df.groupby('Order_month')['Sales'].sum().reset_index()

monthly_sales
```

```python
#Moving average

monthly_sales['Moving_Average'] = monthly_sales['Sales'].rolling(window=3).mean()

#Display
monthly_sales
```

```python
# Plot Sales vs Moving Average

plt.figure(figsize=(10,5))

# Actual Sales

plt.plot(
    monthly_sales['Order_month'],
    monthly_sales['Sales'],
    marker='o',
    label='Actual Sales'
)

# Moving Average

plt.plot(
    monthly_sales['Order_month'],
    monthly_sales['Moving_Average'],
    marker='o',
    linestyle='--',
    label='Moving Average'
)

# Labels

plt.title('Sales Forecasting Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.legend()

plt.grid()

plt.show()
```
