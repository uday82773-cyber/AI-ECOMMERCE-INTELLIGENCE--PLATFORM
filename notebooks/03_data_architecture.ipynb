# Data Architecture & Dataset Planning

## Main Datasets

The project will use multiple datasets to simulate a real-world e-commerce business environment.

### Core Datasets

1. Orders Dataset
   - Order details
   - Sales
   - Profit
   - Quantity
   - Shipping

2. Customers Dataset
   - Customer demographics
   - Segmentation
   - Location

3. Products Dataset
   - Product category
   - Pricing
   - Inventory

4. Reviews Dataset
   - Customer reviews
   - Ratings
   - Sentiment analysis

5. Inventory Dataset
   - Stock levels
   - Warehouse tracking
   - Restocking analysis

---

## Business Architecture

The system will simulate:

Customer → Orders → Products → Revenue → Inventory → Reviews → AI Predictions
